### 练习1：华氏温度转换为摄氏温度。
&nbsp;&nbsp;&nbsp;&nbsp;通过input函数接收任意输入的华氏温度，然后按照提示公式转为摄氏维度，并输出转化后的摄氏温度
> 提示：华氏温度到摄氏温度的转换公式为：$C=(F - 32) \div 1.8$。

```python
#代码

C = lambda F: (F - 32)/1.8
print('摄氏温度为：', round(C(float(input('请输入华氏温度：'))), 2))

```

### 练习2：字符串操作
&nbsp;&nbsp;&nbsp;根据给定字符串，完成如下操作：

* 将字符中出现的“背景”都修改为“北京”
* 判断字符串是否以“。”结尾，如果是输出True，不是则输出False
* 统计字符串中“中国”出现的次数

```python
#代码
s = '北京是中国的首都，具有悠久的历史和璀璨的文化，背景话，又称为京片子，在中国的方言体系中独具特色和魅力，是研究背景历史，乃至中国历史的一大……'

to_replace = '背景'
replace_with = '北京'

while s.find(to_replace) > -1:
    s = s.replace(to_replace, replace_with)

print(s)
print(True if s[-1] == '。' else False)
print('"中国"出现的次数为：', s.count('中国'))

```

### 练习3：条件判断
&nbsp;&nbsp;&nbsp;对于input输入的任意数值x，利用多分支结构实现如下分段函数求值



```python
#代码

x = None
ipt = input('请输入x：')

if ipt.isdigit():
    x = int(ipt)
else:
    try:
        x = float(ipt)
    except ValueError as e:
        print(e, '请输入数字')

if x:
    if x > 1:
        y = 3*x - 5
    elif -1 <= x <= 1:
        y = x + 2
    elif x < -1:
        y = 5*x + 3
    else:
        print('输入值异常')

    print('y = ', round(y, 3))
```

### 练习4：输出任意数n的nn乘法口诀表
&nbsp;&nbsp;&nbsp;&nbsp;通过input函数接收任意输入的整数n，然后再屏幕中输出nn乘法口诀表，例如用户输入9，那么输出九九乘法口诀表如下：


```python
#代码

n = int(input('请输入n：'))

if not 1 <= n <= 9:
    print('输入n范围有误，应为整数1至9')
else:
    for i in range(1, n + 1):
        for m in range(1, i + 1):
            r = m * i
            print(f'{m}*{i} = {r}', end=' ')
        print('')
```

### 练习5：猜数字游戏

1. 随机生成一个数字
> import random  #引入随机数模块<br>
> num = random.randint(1,100)   #随机生成1-100的整数，并赋值给变量num

2. 用户输入一个数，并根据用户输入数据提示
    - 如果用户输入的是数字
        - 如果与随机数相等，则结束游戏
        - 如果大于随机数，提示用户“大了，重新输入”
        - 如果小于随机数，提示用户“小了，重新输入”
    - 如果用户输入的不是数字
        - 如果是exit，则退出游戏
        - 否则，提示用户“请输入数字”

```python
#代码

import random

num = random.randint(1, 100)

while True:

    ipt = input('请输入：')

    if ipt.isdigit():
        ipt = int(ipt)
        
        if ipt != num:
            print('大了，重新输入' if ipt > num else '小了，重新输入')
        else:
            break

    elif 'exit' == ipt:
        raise SystemExit('退出游戏...')
    else:
        print('请输入数字')
    





```

### 练习6：输入一段话，统计每个英文字母（区分大小写）出现的次数，并将统计结果保存在字典中
> 判断字符ch是否为英文字符： if 'A' <= ch <= 'Z' or 'a' <= ch <= 'z':
> "我是Xman，are you ok"这句话的统计结果为：{'X': 1, 'm': 1, 'a': 2, 'n': 1, 'r': 1, 'e': 1, 'y': 1, 'o': 2, 'u': 1, 'k': 1}

```python
#代码

s = input('输入一段话：')

res = {}

for ch in s:
    if 'A' <= ch <= 'Z' or 'a' <= ch <= 'z':
        if ch not in res:
            res.update({ch: 1})
        else:
            res[ch] += 1

print(res)
```
