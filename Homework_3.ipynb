**问题2：基于df_baby_name，使用布尔选择完成如下问题**
* 查找1949年，名字出现最多的10个男性姓名数据
* 查找出名字出现次数大于85000，性别为男性数据
* 查找名字出现次数大于85000，且名字首字母为A的数据
* 查找名字出现次数大于85000，且名字包含an的数据

```python
m_top10 = df_baby_name[(df_baby_name['Sex'] == 'M') & (df_baby_name['Year'] == 1949)].sort_values('Count', ascending=False)['Name'].iloc[:10]

m_name_count_100000 = df_baby_name[(df_baby_name['Count'] > 85000) & (df_baby_name['Sex'] == 'M')]
                                    
a_name_count_100000 = df_baby_name[(df_baby_name['Count'] > 85000) & ([name.startswith('A') for name in df_baby_name['Name']])]

an_name_count_100000 = df_baby_name[(df_baby_name['Count'] > 85000) & (['an' in name for name in df_baby_name['Name']])]

data_in_row3_col1 = df_baby_name.iat[3, 1]

print(m_top10)
print(m_name_count_100000)
print(a_name_count_100000)
print(an_name_count_100000)
print(data_in_row3_col1)
```

## 3. 缺失值处理
**问题1：基于df_weather，完成如下问题：**
* 删除df_weacher中有缺失值的行
* 删除df_weacher中有缺失值的列

<br/>

* 补充：
- isnull():判断是否为缺失值，返回布尔值列表
- fillna(value):用value来填充缺失值
- dropna():删除缺失值
- mean():计算平均值
- 其它方法自行baidu.....

```python
#删除df_weacher中有缺失值的行
df_weather.isnull()
df_weather.dropna(axis='index',how ='any')

#删除df_weacher中有缺失值的列
df_weather.isnull()
df_weather.dropna(axis='columns',how ='any')
```

**问题2：基于df_weather，完成如下问题：**
* 用中位数代替缺失值
* 用均值代替缺失值

```python
# 用中位数代替缺失值
import numpy as np 
for i in range(df_weather['windspeed'].shape[0]):
    if df_weather['windspeed'][i]=='-':
        df_weather['windspeed'][i]=np.nan
df_weather['windspeed']=df_weather['windspeed'].astype('float')
df_weather['windspeed']=df_weather['windspeed'].fillna(df_weather['windspeed'].median())
df_weather['temperature']=df_weather['temperature'].fillna(df_weather['temperature'].median())
df_weather.info()

# 用均值代替缺失值
import numpy as np 
for i in range(df_weather['windspeed'].shape[0]):
    if df_weather['windspeed'][i]=='-':
        df_weather['windspeed'][i]=np.nan
df_weather['windspeed']=df_weather['windspeed'].astype('float')
df_weather['windspeed']=df_weather['windspeed'].fillna(df_weather['windspeed'].mean())
df_weather['temperature']=df_weather['temperature'].fillna(df_weather['temperature'].mean())
df_weather.info()
```
